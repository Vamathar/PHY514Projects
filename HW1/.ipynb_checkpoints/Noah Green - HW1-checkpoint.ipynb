# ODE Integrators

The ODE for a harmonic oscillator is given by Hook's law (in one dimension):
\begin{equation}
F=-kx.
\end{equation}
Putting this in terms of $x$ and its derivatives, we have:
\begin{equation}
x(t)=-\frac{m}{k}\ddot{x}(t)
\end{equation}
The analytical solution for this equation is of course,
\begin{equation}
x(t) = c_1 sin\left(\omega t\right)+c_2 cos\left(\omega t\right)
\end{equation}
where $\omega = \sqrt{\frac{k}{m}}$. The derivative of this equation gives us the velocity,
\begin{equation}
v(t) = c_1\omega cos\left(\omega t\right) - c_2\omega sin\left(\omega t\right).
\end{equation}
Let $x(t=0)=x_0$ and $v(t=0)=v_0$. Then the position and velocity equations become

\begin{equation}
x(t) = \frac{v_0}{\omega} sin\left(\omega t\right)+x_0 cos\left(\omega t\right)
\\
v(t) = v_0 cos\left(\omega t\right) - x_0\omega sin\left(\omega t\right).
\end{equation}

To solve the the problem numerically, I will be using the following coupled first order ordinary differential equations:

\begin{equation}
\frac{dv}{dt} = -\omega^2 x
\\
\frac{dx}{dt} = v.
\end{equation}
This problem is solved below numerically using four different methods. The code for the numerical solutions can be found in *HOint_class.py*. Note that to do a true analysis on the convergence, I would need to make a plot of error vs $\Delta t$. Since this is beyond the scope of the assignment (only different time steps are required), I will simply comment on each algorithm's speed of convergence by looking at the change in error at the same point (I chose the middle point).

```python
import HOint_class as ho
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
# Define parameters of oscillator
period = 2.*math.pi
omega = 1.
x0 = 1.
v0 = 0.

# Define time steps
big_dt = 0.1*period
small_dt = 0.01*period

# Create harmonic oscillator objects
ho_big_dt = ho.HO_int(x0,v0,omega,big_dt,10*period)
ho_small_dt = ho.HO_int(x0,v0,omega,small_dt,10*period)

# Get time coordinates for plotting
time_big = ho_big_dt.time
time_small = ho_small_dt.time

# Analytical solution
ho_analytical_big = ho_big_dt.Analytical()
ho_analytical_small= ho_small_dt.Analytical()

# Function to plot position and velocity comparison
def Compare(time,analytical,numerical,title):
    fig, axes = plt.subplots(nrows=4,ncols=1,figsize=(12,10))
    axes[0].plot(time,numerical[0],'r')
    axes[0].plot(time,analytical[0],'b')
    axes[0].set_title(title,fontsize=16)

    axes[1].plot(time,numerical[1],'r')
    axes[1].plot(time,analytical[1],'b')

    axes[3].set_xlabel('Time (s)')
    axes[0].set_ylabel('Position (m)')
    axes[1].set_ylabel('Velocity (m/s)')
    
    diff_x = list((analytical[0][i]-numerical[0][i]) for i in range(len(time)))
    diff_v = list(analytical[1][i]-numerical[1][i] for i in range(len(time)))
    
    axes[2].plot(time,diff_x,'k')
    axes[2].set_ylabel('Position Difference (m)')
    axes[3].plot(time,diff_v,'k')
    axes[3].set_ylabel('Velocity Difference (m/s)')
    
    plt.show()
    
    max_error_x = max([abs(max(diff_x)),abs(min(diff_x))])
    max_error_v = max([abs(max(diff_v)),abs(min(diff_v))])
    mid_error_x = abs(diff_x[int(len(diff_x)/2)])
    mid_error_v = abs(diff_v[int(len(diff_v)/2)])
    print("{:<15}{:^20}{:^20}".format('','x','v'))
    print("{:<15}{:^20.5f}{:^20.5f}".format('max error (length)',max_error_x,max_error_v))
    print("{:<15}{:^20.5f}{:^20.5f}".format('mid error (length)',mid_error_x,mid_error_v))
```

## Forward Euler Algorithm

```python
# Calculate position and velocity using forward Euler algorithm
ho_big_dt_FE = ho_big_dt.ForwardEuler()
ho_small_dt_FE = ho_small_dt.ForwardEuler()

# Plot results
title = 'Harmonic Oscillator: Forward Euler, $\Delta$t = 0.1T'
Compare(time_big,ho_analytical_big,ho_big_dt_FE,title)

title = 'Harmonic Oscillator: Forward Euler, $\Delta$t = 0.01T'
Compare(time_small,ho_analytical_small,ho_small_dt_FE,title)
```

**Stability**:
    The algorithm obviously unstable. After 10 periods, the numerical solution with $\Delta t = 0.1T$ is about 7 orders of magnitude larger than the analytical solution.
    
**Convergence**:
    The algorithm appears to converge quickly for this function. We get a reduction in error of 4 orders of magnitude by reducing the time step 1 order of magnitude. This is because the error is compounded by the instability of the algorithm due to the oscillatory nature of the solution.

## Backward Euler Algorithm

```python
# Calculate position and velocity using backward Euler algorithm
ho_big_dt_BE = ho_big_dt.BackwardEuler()
ho_small_dt_BE = ho_small_dt.BackwardEuler()

# Plot results
title = 'Harmonic Oscillator: Backward Euler, $\Delta$t = 0.1T'
Compare(time_big,ho_analytical_big,ho_big_dt_BE,title)

title = 'Harmonic Oscillator: Backward Euler, $\Delta$t = 0.01T'
Compare(time_small,ho_analytical_small,ho_small_dt_BE,title)
```

**Stability**:
    The backwards Euler algorithm appears to have the opposite stability problem as the forward Euler algorithm. It appears to always converge to zero given enough iterations. It doesn't diverge to infinity, so technically it is stable.
    
**Convergence**:
    It's harder to compare the convergence on this algorithm, since the max error quickly approaches about 100% for the long $\Delta t$ case. However, it does seem to converge quickly seeing as how the error near 10 seconds goes from nearly 100% for the first case to <40% in the second case.

## RK4 Algorithm

```python
# Calculate position and velocity using 4th order Runge-Kutta Algorithm
ho_big_dt_RK4 = ho_big_dt.RungeKutta()
ho_small_dt_RK4 = ho_small_dt.RungeKutta()

# Plot results
title = 'Harmonic Oscillator: RK4, $\Delta$t = 0.1T'
Compare(time_big,ho_analytical_big,ho_big_dt_RK4,title)

title = 'Harmonic Oscillator: RK4, $\Delta$t = 0.01T'
Compare(time_small,ho_analytical_small,ho_small_dt_RK4,title)
```

**Stability**:
The RK4 method appears slightly unstable. The monotonic increase in the error after 10 periods suggests that the algorithm will diverge to infinity, given enough time.

**Convergence**:
The RK4 algorithm lives up to its name. Reducing $\Delta t$ one order of magnitude reduces the error by four orders of magnitude.

## Leapfrog Method

```python
# Calculate position and velocity using leapfrog method
ho_big_dt_LEAP = ho_big_dt.LeapFrog()
ho_small_dt_LEAP = ho_small_dt.LeapFrog()

# Plot results
title = 'Harmonic Oscillator: Leapfrog Method, $\Delta$t = 0.1T'
Compare(time_big,ho_analytical_big,ho_big_dt_LEAP,title)

title = 'Harmonic Oscillator: Leapfrog Method, $\Delta$t = 0.01T'
Compare(time_small,ho_analytical_small,ho_small_dt_LEAP,title)
```

**Stability**:
This algorithm is stable. The magnitude of the numerical solution stays consistent with the analytical solution. This makes sense since this algorithm has time reversal symmetry, and hence energy is conserved.

**Convergence**:
I will judge this one by looking again at the envelope of the velocity and position difference plots. By reducing the time step by one order of magnitude, we get an improvement of about two orders of magnitude in the error. This makes sense since this algorithm is $O(\Delta t^2)$.


# Shooting

To find the answer analytically, the following system must be solved for $\theta$:
\begin{equation}
x = v_0 cos(\theta)t
\\
y = v_0 sin(\theta)t-\frac{1}{2}gt^2
\end{equation}
When the cannon ball hits the ground at $t=t_f$, $y(t=t_f)=0$. After doing some algebra, it is found that
\begin{equation}
\theta = \frac{1}{2}sin^{-1}\left(\frac{x_fg}{v_0^2}\right).
\end{equation}
Additionally, the time that the ball hits the ground is given by
\begin{equation}
t_f = \frac{2v_0}{g}sin(\theta).
\end{equation}

Below, I find the angle numerically. I did this using the following steps:
* Write a function that determines the trajectory of the cannon ball using forward Euler algorithm.
* Use the trajectory function to write a range function for the cannon. Do this by finding where the height of the trajectory changes sign from positive to negative and find the zero of a line drawn between the points at that crossover.
* Find $\theta$ using the midpoint method. I will be finding the root of the function $f(\theta) = range(\theta)-distance(target)$.

```python
def analytical_height(v0,theta,dt):
    """
    Analytically calculates height of cannon ball until it hits the gound.
    
    Input:
        v0: float, initial velocity
        theta: angle of cannon wrt horizontal
        dt: time step
        
    Returns:list(list(float),list(float)), [[time],[height]]
    """
    g = 9.81
    ylist = list()
    tlist = list()
    y = 0.
    t = 0.
    ylist.append(y)
    tlist.append(t)
    while y >= 0.:
        t += dt
        y = v0*math.sin(theta)*t-0.5*g*t**2
        ylist.append(y)
        tlist.append(t)
    return [tlist,ylist]
```

```python
def analytical_theta(distance,v0):
    return 0.5*math.asin(distance*9.81/v0**2)
def analytical_range(v0,theta):
    return v0**2*math.sin(2.*theta)/9.81
```

```python
def numerical_height(v0,theta,dt):
    """
    Numerically calculates height of cannon ball until it hits the gound using forward Euler algorithm.
    
    Input:
        v0: float, initial velocity
        theta: angle of cannon wrt horizontal
        dt: time step
        
    Returns:list(list(float),list(float)), [[time],[height]]
    """
    g = 9.81
    ylist = list()
    tlist = list()
    vy = v0*math.sin(theta)
    y = 0.
    t = 0.
    tlist.append(t)
    ylist.append(y)
    while y >= 0.:
        t+=dt
        y+=vy*dt
        tlist.append(t)
        ylist.append(y)
        vy-=g*dt
    return [tlist,ylist]        
```

```python
def numerical_range(v0,theta,dt):
    """
    Numerically calculates the range of the cannon using forward Euler algorithm.

    Input:
        v0: float, initial velocity
        theta: angle of cannon wrt horizontal
        dt: time step 

    Returns: float, range of cannon
    """
    # Calculate height wrt time
    tlist,ylist = numerical_height(v0,theta,dt)
    # calculate horizontal velocity
    vx = v0*math.cos(theta)
    # calculate distance cannon ball travelled
    x1 = vx*tlist[-2]
    x2 = vx*tlist[-1]
    y1 = ylist[-2]
    y2 = ylist[-1]
    slope = (y2-y1)/(x2-x1)
    return x1-y1/slope    
```

```python
def numerical_theta(distance,v0,dt,max_iter,error):
    """
    Determines angle of cannon using midpoint method.
    
    Input: 
        distance: float, distance to target
        v0: float, initial velocity of cannon ball
        dt: float, time step
        max_iter: integer, limit number of iterations to protect against infinite loops
        error: float, size of error interval around actual distance
        
    Returns: float, theta to achieve distance
    """
    # max range is at 45 degrees
    th0 = 0.
    th1 = math.pi/4.
    th2 = 0.
    dth0 = numerical_range(v0,th0,dt) - distance
    dth1 = numerical_range(v0,th1,dt) - distance
    diff = dth1-dth0
    i = 0
    while (abs(diff) > error) and i < max_iter:
        i+=1
        dth0 = numerical_range(v0,th0,dt) - distance
        dth1 = numerical_range(v0,th1,dt) - distance
        diff = dth1-dth0
        th2 = (th0 + th1)/2.
        dth2 = numerical_range(v0,th2,dt) - distance
        pos0 = dth0 > 0
        pos2 = dth2 > 0
        if (pos0 and pos2) or ((not pos0) and (not pos2)):
            th0 = th2
        else:
            th1 = th2
    if i >= max_iter:
        print("Warning: function numerical_theta did not converge.")
        return th2
    return th2
```

```python
# Boundary conditions, etc.
distance = 1500.
v0 = 150.
dt = 0.1
max_iter = 10000
error = 10.

# Find theta analytically and numerically
atheta = analytical_theta(distance,v0)
ntheta = numerical_theta(distance,v0,dt,max_iter,error)
# Make trajectory lists
atime,ay = analytical_height(v0,atheta,dt)
ntime,ny = numerical_height(v0,ntheta,dt)
ax = [v0*math.cos(atheta)*time for time in atime]
nx = [v0*math.cos(atheta)*time for time in ntime]
# Find difference between trajectories for comparison
dy = [ay[i]-ny[i] for i in range(len(ay))]

# Plot results
fig,axes = plt.subplots(figsize = (12,8))
axes.plot(ax,ay,'b',label='Analytical Trajectory')
axes.plot(nx,ny,'r',label='Numerical Trajectory')
axes.set_xlabel('x (m)')
axes.set_ylabel('y (m)')
axes.set_title('Trajectory of Cannon Ball',fontsize=16)
plt.legend()
plt.show()

dfig,daxes = plt.subplots(figsize = (12,8))
daxes.plot(ax,dy,'k')
daxes.set_xlabel('x (m)')
daxes.set_ylabel('$\Delta$y (m)')
daxes.set_title('Difference Between Analytical and Numerical Trajectories',fontsize=16)
plt.show()

# Print results
print("{:<10}{:^12}{:^12}".format('','Analytical','Numerical'))
print("{:<10}{:^12.2f}{:^12.2f}".format('theta(deg)',atheta*180/math.pi,ntheta*180/math.pi))
```
